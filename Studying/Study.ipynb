## Some gradient decent exersie

## <center> $ \textbf{w}^k =\textbf{w}^{k-1} - \alpha \nabla g(\textbf{w}^{k-1}) $ <center> 

```python
def f(x):
    return x**3-12*x**2

def fd(x):
    return 3*x**2 -24*x

def fdd(x):
    return 3*x-24



t1 = np.arange(-1.5, 10, 0.01)
plt.figure()
plt.figure(figsize=(15, 12))
plt.subplot(211)

plt.plot(t1, f(t1), 'k')


w = 1
step = 0.01
i = 0
while i<30:
    plt.plot(w,f(w),'bo')
    w = w -step*fd(w) #Formule is there
    print(fd(w))
    if abs(fd(w)) < 10**-1 :
        break
    i += 1
    

```

# Derivative of $ g(\textbf{w}) = \begin{bmatrix} 0.4 & 9 \end{bmatrix} w^2 + 0.25 w^T w+ sum(w) $

 $g(\textbf{w}) = \begin{bmatrix} 0.4 & 9 \end{bmatrix} {\begin{bmatrix} w_0 \\ w_1 \end{bmatrix}}^2 + 0.25 {\begin{bmatrix} w_0 \\ w_1 \end{bmatrix}}^T {\begin{bmatrix} w_0 \\ w_1 \end{bmatrix}}+ w_0 + w_1$ 

 $g(\textbf{w}) = \begin{bmatrix} 0.4 & 9 \end{bmatrix} {\begin{bmatrix} w_0 \\ w_1 \end{bmatrix}}^2 + 0.25 {\begin{bmatrix} w_0 & w_1 \end{bmatrix}} {\begin{bmatrix} w_0 \\ w_1 \end{bmatrix}}+ w_0 + w_1$ 

 $g(\textbf{w}) = \begin{bmatrix} 0.4 & 9 \end{bmatrix} {\begin{bmatrix} w_0 \\ w_1 \end{bmatrix}}{\begin{bmatrix} w_0 \\ w_1 \end{bmatrix}} + 0.25w_0^2 + 0.25w_1^2 + w_0 + w_1$ 

 $g(\textbf{w}) = {\begin{bmatrix} w_0(0.4w_0 + 9w_1) \\ w_1(0.4w_0 + 9w_1) \end{bmatrix}} + 0.25w_0^2 + 0.25w_1^2 + w_0 + w_1$ 

 $g(\textbf{w}) = {\begin{bmatrix}0.4w_0^2 + 9w_1w_0 \\ 0.4w_1w_0 + 9w_1^2 \end{bmatrix}} + 0.25w_0^2 + 0.25w_1^2 + w_0 + w_1$

## $$\nabla g(\textbf{w}) = \begin{bmatrix} \frac{\partial g}{\partial \omega_{0}} \\ \frac{\partial g}{\partial \omega_{1}}\end{bmatrix}$$

 $g(\textbf{w}) = {\begin{bmatrix}0.8w_0 + 9w_1  + 0.5w_0  + 1\\ 0.4w_0 + 18w_1  + 0.5w_1 + 1 \end{bmatrix}}$ 

```python

A = np.array([[0.4], [9]])

print(A[1])
```

```python
def g(x):
    result = np.matmul(np.array([0.4, 9]),np.power(x,2)) + 0.25*np.matmul(x.transpose(),x) + np.sum(x) #cost function
    return result[0][0]
def grad_w(w):
    result = np.array([[0.8*w[0][0]+9*w[1][0]+0.5*w[0][0]+1],[0.4*w[0][0]+18*w[1][0]+0.5*w[1][0]+1]])
    return result# gradient function
```
