## Some gradient decent exersie

## <center> $ \textbf{w}^k =\textbf{w}^{k-1} - \alpha \nabla g(\textbf{w}^{k-1}) $ <center> 

```python
def f(x):
    return x**3-12*x**2

def fd(x):
    return 3*x**2 -24*x

def fdd(x):
    return 6*x-24



t1 = np.arange(-1.5, 10, 0.01)
plt.figure()
plt.figure(figsize=(15, 12))
plt.subplot(211)

plt.plot(t1, f(t1), 'k')


w = 1
step = 0.01
i = 0
while i<30:
    plt.plot(w,f(w),'bo')
    w = w -step*fd(w) #Formule is there
    print(fd(w))
    if abs(fd(w)) < 10**-1 :
        break
    i += 1
    

```

### <center> $ g(\textbf{w}) = 0.26(\textbf{w}^T\textbf{w}) + 0.48(sum(\textbf{w}))-cos(\textbf{w}^T\textbf{w})$ <center>

$$ g(\textbf{w}) = 0.26(\textbf{w}^T\textbf{w}) + 0.48(sum(\textbf{w}))-cos(\textbf{w}^T\textbf{w})$$

$$g(\textbf{w}) = 0.26{\begin{bmatrix}w_0 & w_1 \end{bmatrix}} {\begin{bmatrix}w_0 \\ w_1 \end{bmatrix}} + 0.48w_0 + 0.48w_1 - \cos{w^Tw}$$

$$g(\textbf{w}) = 0.26 \left(w_0^2 +  w_0 w_1  + w_1 w_0 + w_1^2\right) + 0.48w_0 + 0.48w_1 - \cos{\left(w_0^2 +  w_0 w_1  + w_1 w_0 + w_1^2\right)}$$

$$g(\textbf{w}) = 0.26 w_0^2 +  0.52w_1 w_0 + 0.26w_1^2 + 0.48w_0 + 0.48w_1 - \cos{\left(w_0^2 + 2w_1 w_0 + w_1^2\right)}$$

## $$\nabla g(\textbf{w}) = \begin{bmatrix} \frac{\partial g}{\partial \omega_{0}} \\ \frac{\partial g}{\partial \omega_{1}}\end{bmatrix}$$

$$\nabla g(\textbf{w}) = \begin{bmatrix} 0.52w_0+0.52w_1 + 0.48 \\ \frac{\partial g}{\partial \omega_{1}}\end{bmatrix} $$

$$\nabla g(\textbf{w}) =\begin{bmatrix} 0.52w_0+0.52w_1 + 0.48 -2(w_0+w_1) \sin{\left((w_0+w_1)^{2}\right)} \\ 0.52w_0+0.52w_1 + 0.48 -2(w_0+w_1) \sin{\left((w_0+w_1)^{2}\right)}\end{bmatrix} $$

```python
def grad_w_1st(w):
    result = w[0][0] * 0.52 +  w[1][0] * 0.52 + 0.48 - 2*(w[0][0]+w[1][0])*np.sin((w[0][0]+w[1][0])**2)
    return result

def grad_w_1st2(w):
    result = 0.52 * np.sum(w) + 0.48 - 2*np.sum(w)*np.sin(np.sum(w)**2)
    return np.array([[result],[result]])

def grad_w_2nd(w):
    sumW = np.sum(w)
    result = -2*np.sin(sumW**2)-4*(sumW**2)*np.cos(sumW**2)+51
    return np.array([[result],[result]])

print(grad_w_1st(A))
print(grad_w_1st2(A))
print(grad_w_2nd(A))
```
