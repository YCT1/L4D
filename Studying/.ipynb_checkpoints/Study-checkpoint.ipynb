## Some gradient decent exersie

```python
def f(x):
    return x**3-12*x**2

def fd(x):
    return 3*x**2 -24*x

def fdd(x):
    return 3*x-24



def f2(x):
    return 3*x**4/4 + 2*x**3/2

def f2d(x):
    return 3*x**3+2*x**2

def f2dd(x):
    return 9*x**2 + 4*x

t1 = np.arange(-1.5, 1, 0.01)
plt.figure()
plt.figure(figsize=(15, 100))
plt.subplot(211)

plt.plot(t1, f2(t1), 'k')


w = 1
step = 0.05
while True:
    print(f2dd(w))
    plt.plot(w,f2(w),'bo')
    w+=step*-np.sign(f2d(w))
    if abs(f2d(w)) < 10**-5 :
        break
    

```
