Before we continue let's write Naive Bayes equation as follows;

$$P\left(class|data\right)=\frac{P\left(class|data\right) \cdot P\left(class\right)}{P\left(data\right)}$$

So, if we do necacary maths we get this
$$g_{i}(x)=-\frac{1}{2}\left(x-\mu_{i}\right)^{t} \Sigma_i^{-1}\left(x-\mu_{i}\right)-\frac{1}{2} \ln \left|\Sigma_{i}\right|+\ln P\left(c_{i}\right)$$

However, we can divide this equation to two part where $\Sigma_{1}=\Sigma_{2}$

$$g_{i}(x)=-\frac{1}{2}\left(x-\mu_{i}\right)^{t} \Sigma_i^{-1}\left(x-\mu_{i}\right)+\ln P\left(c_{i}\right)$$

We are going to use this function where $\Sigma_{1}=\Sigma_{2}$ and if cov matrixes are not equal each others, we will use general equation.


## Lets create general function to calculate the $g_i(x)$

```python
### I am going to implement with two function, first one will find mean vectors, cov matrixes and p(x) return them
### Other function will calculate and return result of g(x) for class 0 and class 1


#This function finds cov matrixes and mean vector for desired dataset
def finder(trainingSamples):
    #First lets convert data to numpy
    df1 = trainingSamples.values
    df1_x1 = trainingSamples['x1'].values
    df1_x2 = trainingSamples['x2'].values
    
    #Let's partition to classes we assume there are only two classes where class 0 = 0 and class 1 = 1
    
    df1_class0 = trainingSamples[trainingSamples['y']==0]
    df1_class1 = trainingSamples[trainingSamples['y']==1]
    
    
    #Let's divide more and spearete x1 and x2 for each class
    
    df1_class0_x1 = df1_class0['x1'].values
    df1_class0_x2 = df1_class0['x2'].values
    
    df1_class1_x1 = df1_class1['x1'].values
    df1_class1_x2 = df1_class1['x2'].values
    
    
    #Lets find covariance matrixes
    
    data = np.array([df1_class0_x1 ,df1_class0_x2])
    covMatrix_df1_class0 = np.cov(data,bias=True)
    
    data = np.array([df1_class1_x1 ,df1_class1_x2])
    covMatrix_df1_class1 = np.cov(data,bias=True)
    
    
    
    #Let's find mean vectors
    
    df1_class0_mean = np.mean(df1_class0 , axis=0)
    df1_class1_mean = np.mean(df1_class1 , axis=0)

    df1_class0_mean_vector = np.array([[df1_class0_mean['x1'],df1_class0_mean['x2']]])
    df1_class0_mean_vector = np.transpose(df1_class0_mean_vector)

    df1_class1_mean_vector =np.array([[df1_class1_mean['x1'],df1_class1_mean['x2']]])
    df1_class1_mean_vector = np.transpose(df1_class1_mean_vector)
    
    
    #Probabilty
    pci0 = df1_class0.size/df1.size
    pci1 = df1_class1.size/df1.size
    #We have find the means vectors now we can write the formulas,
    return df1_class0_mean_vector,covMatrix_df1_class0, df1_class1_mean_vector, covMatrix_df1_class1,pci0,pci1



    

#vc0 = mean vector class 0
#vc1 = mean vector class 1
#cmc0 = cov Matrix class 0
#cmc1 = cov Matrix class 1
#pci0 = p(c_0) for class 0
#pci1 = p(c_0) for class 1
def g_iX(x,vc0,vc1,cmc0,cmc1,pci0,pci1):
    #In this function we get necacary data and x variable and create linear discriminant function
    if np.allclose(cmc0,cmc1):
        #If the cov matrixes are equal, we can use special equation
        result_class0 = (-1/2) * np.matmul(np.matmul(np.transpose(x-vc0),np.linalg.inv(cmc0)),(x-vc0)) + math.log(pci0)
        result_class1 = (-1/2) * np.matmul(np.matmul(np.transpose(x-vc1),np.linalg.inv(cmc1)),(x-vc1)) + math.log(pci1)
        return result_class0,result_class1
    else:
        #We are going to implement general equation
        result_class0 = (-1/2)*(np.matmul(np.matmul(np.transpose(x-vc0),np.linalg.inv(cmc0)),(x-vc0)))-(1/2)*math.log(np.linalg.det(cmc0))+math.log(pci0)
        result_class1 = (-1/2)*(np.matmul(np.matmul(np.transpose(x-vc1),np.linalg.inv(cmc1)),(x-vc1)))-(1/2)*math.log(np.linalg.det(cmc1))+math.log(pci1)
        return result_class0,result_class1



#Lets check for DATASET 1

df1_x1 = df1['x1'].values
df1_x2 = df1['x2'].values

#Let's partition to classes we assume there are only two classes where class 0 = 0 and class 1 = 1

df1_class0 = df1[df1['y']==0].values
df1_class1 = df1[df1['y']==1].values

#Let's get the test dataset 1
tst1 = pd.read_csv("test1.csv") 
tst1_class0 = tst1[tst1['y']==0].values
tst1_class1 = tst1[tst1['y']==1].values

df1_class0_mean_vector,covMatrix_df1_class0, df1_class1_mean_vector, covMatrix_df1_class1,pci0,pci1 = finder(df1)
df2_class0_mean_vector,covMatrix_df2_class0, df2_class1_mean_vector, covMatrix_df2_class1,pci0,pci1 = finder(df2)
print("")
print("DATASET 1, CLASS 0")
print("")
error_number_dataset1 = 0
i = 0
while i < tst1_class0.shape[0]:
    x = np.array([[tst1_class0[i][0]],[tst1_class0[i][1]]])
    result_class0,result_class1 = g_iX(x,df1_class0_mean_vector,df1_class1_mean_vector,covMatrix_df1_class0,covMatrix_df1_class1,pci0,pci1)

    if result_class0 > result_class1:
        print("TRUE: Test Set 0 Class 0")
    else:
        print("FALSE: Test Set 0 Class 0")
        error_number_dataset1 += 1
    i += 1

print("")
print("DATASET 1, CLASS 1")
print("")

i = 0   
while i < tst1_class1.shape[0]:
    x = np.array([[tst1_class1[i][0]],[tst1_class1[i][1]]])
    result_class0,result_class1 = g_iX(x,df1_class0_mean_vector,df1_class1_mean_vector,covMatrix_df1_class0,covMatrix_df1_class1,pci0,pci1)

    if result_class1 > result_class0:
        print("TRUE: Test Set 0 Class 1")
    else:
        print("FALSE: Test Set 0 Class 1")
        error_number_dataset1 += 1
    i += 1
    

    
#Lets check for DATASET 2

df2_x1 = df2['x1'].values
df2_x2 = df2['x2'].values

#Let's partition to classes we assume there are only two classes where class 0 = 0 and class 1 = 1

df2_class0 = df2[df2['y']==0].values
df2_class1 = df2[df2['y']==1].values


tst2 = pd.read_csv("test2.csv") 
tst2_class0 = tst2[tst2['y']==0].values
tst2_class1 = tst2[tst2['y']==1].values


print("")
print("DATASET 2, CLASS 0")
print("")
error_number_dataset2 = 0
i = 0
while i < tst2_class0.shape[0]:
    x = np.array([[tst2_class0[i][0]],[tst2_class0[i][1]]])
    result_class0,result_class1 = g_iX(x,df2_class0_mean_vector,df2_class1_mean_vector,covMatrix_df2_class0,covMatrix_df2_class1,pci0,pci1)

    if result_class0 > result_class1:
        print("TRUE: Test Set 2 Class 0")
    else:
        print("FALSE: Test Set 2 Class 0")
        error_number_dataset2 += 1
    i += 1

print("")
print("DATASET 2, CLASS 1")
print("")

i = 0   
while i < tst2_class1.shape[0]:
    x = np.array([[tst2_class1[i][0]],[tst2_class1[i][1]]])
    result_class0,result_class1 = g_iX(x,df2_class0_mean_vector,df2_class1_mean_vector,covMatrix_df2_class0,covMatrix_df2_class1,pci0,pci1)

    if result_class1 > result_class0:
        print("TRUE: Test Set 2 Class 1")
    else:
        print("FALSE: Test Set 2 Class 1")
        error_number_dataset2 += 1
    i += 1
    
```
